```python
import yfinance as yf
stock_list = ['GME', 'AMC', 'CLOV']
msft = yf.Ticker("GME AMC CLOV")

# get stock info
print(msft.info)

# get historical market data
hist = msft.history(interval='30m', period="1d", start='2021-09-14', end='2021-09-15')
```

```python
hist
```

```python
df = yf.download(tickers='GME AMC CLOV',
                 interval='30m',
                 start='2021-08-14', 
                 end='2021-09-17',
                 group_by='ticker',
                 prepost=True)  #prepost=True for pre- and post-market hours

df.head()
```
